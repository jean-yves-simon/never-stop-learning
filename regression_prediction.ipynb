# Étape 1 - Importer les modules 

```python
from tensorflow import keras 

from keras.datasets import boston_housing 
from keras.models import Sequential 
from keras.layers import Dense 

from keras.callbacks import EarlyStopping 
from sklearn import preprocessing 
from sklearn.preprocessing import scale
```

# Étape 2 - Charger les données 

```python
(x_train, y_train), (x_test, y_test) = boston_housing.load_data()
```

boston_housing est un jeu de données fourni par Keras. Il représente une collection d'informations sur le logement dans la région de Boston, chacune ayant 13 caractéristiques.

# Étape 3 − Traiter les données

Modifions l'ensemble de données en fonction de notre modèle, afin que nous puissions alimenter notre modèle. Les données peuvent être modifiées en utilisant le code ci-dessous -



```python
x_train_scaled = preprocessing.scale(x_train) 
scaler = preprocessing.StandardScaler().fit(x_train) 
x_test_scaled = scaler.transform(x_test)
```

Ici, nous avons normalisé les données d'entraînement à l'aide de la fonction sklearn.preprocessing.scale . La fonction preprocessing.StandardScaler().fit renvoie un scalaire avec la moyenne normalisée et l'écart type des données d'apprentissage, que nous pouvons appliquer aux données de test à l'aide de la fonction scalar.transform . Cela normalisera également les données de test avec le même réglage que celui des données de formation.

# Étape 4 - Créer le modèle

Créons le modèle réel. 

```python
model = Sequential() 
model.add(Dense(64, kernel_initializer = 'normal', activation = 'relu',input_shape = (13,))) 
model.add(Dense(64, activation = 'relu')),
model.add(Dense(1))
```

# Étape 5 - Compiler le modèle

Compilons le modèle en utilisant la fonction de perte, l'optimiseur et les métriques sélectionnés. 

```python
model.compile(
   loss = 'mse', 
   optimizer = 'RMSprop', 
   metrics = ['mean_absolute_error']
)
```

# Étape 6 - Former le modèle

Entraînons le modèle en utilisant la méthode fit() . 

```python
history = model.fit(
   x_train_scaled, y_train,    
   batch_size=128, 
   epochs = 500, 
   verbose = 1, 
   validation_split = 0.2, 
   callbacks = [EarlyStopping(monitor = 'val_loss', patience = 20)]
)
```

Ici, nous avons utilisé la fonction de rappel, EarlyStopping . Le but de ce rappel est de surveiller la valeur de perte à chaque époque et de la comparer à la valeur de perte de l'époque précédente pour trouver l'amélioration de l'entraînement. S'il n'y a pas d'amélioration pour les temps de patience , alors tout le processus sera arrêté.

# Étape 7 − Évaluer le modèle

Évaluons le modèle à l'aide de données de test. 

```python
score = model.evaluate(x_test_scaled, y_test, verbose = 0) 
print('Test loss:', score[0]) 
print('Test accuracy:', score[1])
```

# Étape 8 - Prédire

Enfin, prédisez en utilisant les données de test comme ci-dessous - 

```python
prediction = model.predict(x_test_scaled) 
print(prediction.flatten()) 
print(y_test)
```

La sortie des deux tableaux a une différence d'environ 10 à 30 % et cela indique que notre modèle prédit avec une plage raisonnable.
